```python
# from the pandas data frame
batsman = pd.read_csv("DataSets/sharma-kohli.csv")
batsman
```

```python
# title and label

plt.plot(batsman["index"], batsman["V Kohli"])
plt.plot(batsman["index"], batsman["RG Sharma"])

plt.title("Rohit sharma vs virat kohli compatision")
plt.xlabel("Season")
plt.ylabel("Run Score")
```

```python
# colors(hex)
plt.plot(batsman["index"], batsman["V Kohli"], color="green")
plt.plot(batsman["index"], batsman["RG Sharma"], color="red")

plt.title("Rohit sharma vs virat kohli compatision")
plt.xlabel("Season")
plt.ylabel("Run Score")
```

```python
# line style and width
plt.plot(
    batsman["index"], batsman["V Kohli"], color="green", linestyle="solid", linewidth=3
)
plt.plot(batsman["index"], batsman["RG Sharma"], color="red", linestyle="dashdot")


plt.title("Rohit sharma vs virat kohli compatision")
plt.xlabel("Season")
plt.ylabel("Run Score")
plt.grid()
```

```python
# marker

plt.plot(
    batsman["index"],
    batsman["V Kohli"],
    color="green",
    linestyle="solid",
    linewidth=3,
    marker="D",
    markersize=10,
    label="virat",
)
plt.plot(
    batsman["index"],
    batsman["RG Sharma"],
    color="red",
    linestyle="dashdot",
    marker="o",
    markersize=5,
    label="Rohit sharma",
)


plt.title("Rohit sharma vs virat kohli compatision")
plt.xlabel("Season")
plt.ylabel("Run Score")
plt.legend(
    loc="best"
)  # loc parameter automatically decide kari ley ke graph ma kya place khali che tya legend muki aapde aapde value joei to chane pan kari shaki

# best
# upper right
# upper left
# lower rigth
# lower left
# right
# center left
# center right
# center right
# center left
```

```python
df = pd.read_csv("DataSets/batsman_season_record.csv")
df
```

```python
plt.bar(df["batsman"], df["2015"])
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df["2015"], width=0.2)
plt.bar(np.arange(df.shape[0]), df["2016"], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df["2017"], width=0.2)

plt.xticks(np.arange(df.shape[0]), df["batsman"])
plt.show()
```

```python
childern = [10, 20, 40, 10, 30]
color = [
    "red red red red red",
    "blue blue blue blue blue",
    " green green green green",
    " yellow yellow yellow",
    " pink pink pink pink",
]

plt.bar(color, childern)
plt.xticks(rotation="vertical")
```

```python
# stacked bar chart
plt.bar(df["batsman"], df["2015"], label="2015")
plt.bar(df["batsman"], df["2016"], bottom=df["2015"], label="2016")
plt.bar(df["batsman"], df["2017"], bottom=(df["2015"] + df["2016"]), label="2015")
plt.legend()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Sample data
x_data = [1, 2, 3, 4, 5, 6, 7]
y_data = [10, 11, 8, 15, 13, 20, 17]

# Create a trace for the line chart
trace = go.Scatter(x=x_data, y=y_data, mode="lines", name="Line Chart")

# Create a layout for the plot
layout = go.Layout(
    title="Simple Line Chart", xaxis=dict(title="X-axis"), yaxis=dict(title="Y-axis")
)

# Create a figure and add the trace and layout
fig = go.Figure(data=[trace], layout=layout)

# Save the plot as an HTML file
fig.write_html("line_chart.html")
```
