### subplots

```python
# a diff way to plot graphs
batters = pd.read_csv("DataSets/batter.csv")
batters
```

```python
plt.scatter(batters["avg"], batters["strike_rate"])
```

```python
fig, ax = plt.subplots()

ax.scatter(batters["avg"], batters["strike_rate"], color="red", marker="+")
ax.set_title("something")
ax.set_xlabel("Avg")
ax.set_ylabel("Strike rate")
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10, 6))

ax[0].scatter(batters["avg"], batters["strike_rate"], color="red")
ax[1].scatter(batters["avg"], batters["runs"])


ax[0].set_title("Avg Vs Strike rate")
ax[0].set_ylabel("strike rate")
ax[1].set_ylabel("runs")
ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel("avg")
```

```python
ax
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 6))

ax[0, 0].scatter(batters["avg"], batters["strike_rate"], color="red")
ax[0, 1].scatter(batters["avg"], batters["runs"])
ax[1, 0].hist(batters["avg"])
ax[1, 1].hist(batters["runs"])

ax[0].set_title("Avg Vs Strike rate")
ax[0].set_ylabel("strike rate")
ax[1].set_ylabel("runs")
# ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel("avg")
```

```python
# third way
fig = plt.figure()

ax1 = fig.add_subplot(211)  # 2 rows 1 columns 1 graph
ax1.scatter(batters["avg"], batters["strike_rate"], color="red")

ax2 = fig.add_subplot(212)  # 2 rows 1 columns 2 graph
ax2.scatter(batters["avg"], batters["runs"])
```

## 3D surface plot

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
```

```python
xx, yy = np.meshgrid(x, y)
xx.shape
```

```python
xx
```

```python
yy
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection="3d")
p = ax.plot_surface(xx, yy, z, cmap="viridis")
fig.colorbar(p)
```

## heatmap

```python
dev = pd.read_csv("DataSets/IPL_Ball_by_Ball_2008_2022.csv")
dev.head()
```

```python
temp = dev[(dev["ballnumber"].isin([1, 2, 3, 4, 5, 6])) & (dev["batsman_run"] == 6)]
temp
```

```python
grid = temp.pivot_table(
    index="overs", columns="ballnumber", values="batsman_run", aggfunc="count"
)
grid
```

```python
plt.figure(figsize=(30, 8))
plt.imshow(grid)
plt.colorbar()
```
